```python
import pandas as pd
import ast
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('movies_new.csv')
```

```python
movies['genres'] = movies['genres'].apply(ast.literal_eval)
```

```python
exploded_row = movies.explode('genres')
```

```python
exploded_row
```

```python
new_gen = exploded_row.groupby('genres').mean()['net'].reset_index()
```

```python
sns.barplot(x='genres', y='net', data=new_gen ,palette='Set1')
plt.xticks(rotation=90)

plt.xlabel('Genre')
plt.ylabel('Net profit')
plt.title('Net profit per genre')
plt.savefig('top10gen.png', bbox_inches='tight')

plt.show()
```
