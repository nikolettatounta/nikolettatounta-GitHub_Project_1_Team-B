```python
import pandas as pd
import datetime as dt
import json
import ast
```

```python
movies = pd.read_csv('movies_metadata.csv')
```

```python
movies.head()
```

```python
movies.shape
```

```python
movies.info()
```

drop some columns! belongs_to_collection, homepage, poster_path, spoken_languages, tagline. revisited movies without useless columns

```python
c_todrop = ['belongs_to_collection', 'homepage', 'poster_path', 'spoken_languages', 'tagline']
movies = movies.drop(columns = c_todrop)
```

```python
movies.shape
```

```python
movies.info()
```

dropna. reviseted movies without nas

```python
movies = movies.dropna().reset_index()
```

```python
movies.shape
```

conversion! budget(nm), release_date(dt)

```python
movies['budget'] = pd.to_numeric(movies['budget'])
movies['release_date'] = pd.to_datetime(movies['release_date'])
```

```python
movies.info()
```

clean the genre column to just the name

```python
movies['genres'] = movies['genres'].apply(ast.literal_eval)
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [entry['name'] for entry in x])
```
